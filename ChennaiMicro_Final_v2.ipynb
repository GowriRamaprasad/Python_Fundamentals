```python
import string
from nltk import tokenize as tk
allwords=[]
def extract_words(mystring):
    mystring=mystring.translate(str.maketrans(string.punctuation,'                                '))
    mystring=tk.word_tokenize(mystring)
    return mystring
for tweet in tweets:
    tweet['words']=extract_words(tweet['text'])
    for word in tweet['words']:
        allwords.append(word)
allwords = [word for word in allwords if len(word)>2]

# load nltk's English stopwords as variable called 'stopwords'
stopwords = nltk.corpus.stopwords.words('english')

nostopwords = [word for word in allwords if word not in stopwords]
print(Counter(nostopwords))
```

```python
#!/usr/bin/env python
"""
Masked wordcloud
================
Using a mask you can generate wordclouds in arbitrary shapes.
"""

from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS

d = path.dirname('/Users/gowriramaprasad/Desktop/')

# Read the whole text.
text = open(path.join(d, 'cleaned_tweet_text.txt')).read()

# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "chennai_map.png")))

stopwords = set(STOPWORDS)

wc = WordCloud(background_color="white", max_words=25, mask=alice_mask,
               stopwords=stopwords, max_font_size = 70)
# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "chennai_map.png"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
